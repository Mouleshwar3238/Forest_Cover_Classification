# Exploratory Data Analysis - Forest Cover Classification

```python
# Importing the required libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from matplotlib.ticker import ScalarFormatter
```

```python
# Reading the dataset
df = pd.read_csv('covtype.csv')
df.shape
```

```python
df.head()
```

```python
df.describe()
```

```python
# Checking for missing values
print("Number of missing values in the dataset:",df.isnull().sum())
```

## Exploratory Data Analysis

```python
# Visualizing the distribution of the target variable
plt.figure(figsize=(16,8))
sns.countplot(x=df['Cover_Type'], data=df, width=0.5)
plt.title('Distribution of the Target Variable',color='black', fontsize=16)
plt.xlabel('Class Label (Cover Type)',color='black', fontsize=14)
plt.ylabel('Count',color='black', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid(True, linewidth=1, color='black', alpha=0.2)
plt.show()

per_cl1 = round(len(df[df['Cover_Type'] == 1])*100 / len(df), 3) 
per_cl2 = round(len(df[df['Cover_Type'] == 2])*100 / len(df), 3) 
per_cl3 = round(len(df[df['Cover_Type'] == 3])*100 / len(df), 3) 
per_cl4 = round(len(df[df['Cover_Type'] == 4])*100 / len(df), 3) 
per_cl5 = round(len(df[df['Cover_Type'] == 5])*100 / len(df), 3) 
per_cl6 = round(len(df[df['Cover_Type'] == 6])*100 / len(df), 3) 
per_cl7 = round(len(df[df['Cover_Type'] == 7])*100 / len(df), 3)

print('Number of Samples:')
print('Class 1:',len(df[df['Cover_Type'] == 1]),'('+str(per_cl1)+' %)')
print('Class 2:',len(df[df['Cover_Type'] == 2]),'('+str(per_cl2)+' %)')
print('Class 3:',len(df[df['Cover_Type'] == 3]),'('+str(per_cl3)+' %)')
print('Class 4:',len(df[df['Cover_Type'] == 4]),'('+str(per_cl4)+' %)')
print('Class 5:',len(df[df['Cover_Type'] == 5]),'('+str(per_cl5)+' %)')
print('Class 6:',len(df[df['Cover_Type'] == 6]),'('+str(per_cl6)+' %)')
print('Class 7:',len(df[df['Cover_Type'] == 7]),'('+str(per_cl7)+' %)')
```

```python
cover_type_noise = df['Cover_Type'] + np.random.uniform(-0.25, 0.25, size=len(df))

plt.figure(figsize=(16, 8))
sns.histplot(cover_type_noise, bins=75, kde=True, edgecolor='black', alpha=0.5)

plt.title('Skewed Distribution Curve', color='black', fontsize=16)
plt.xlabel('Class Label (Cover Type)', color='black', fontsize=14)
plt.ylabel('Count', color='black', fontsize=14)

ax = plt.gca()
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.ticklabel_format(style='plain', axis='y')

plt.tick_params(axis='both', which='major', labelsize=14)
plt.grid(True, linewidth=1, color='black', alpha=0.2)

plt.show()
```

```python
# Computing and plotting the correlation matrix
corr = df.corr()
plt.figure(figsize=(20, 16))
sns.heatmap(corr, annot=False, fmt=".2f", cmap='coolwarm', linewidths=0.5, linecolor='black')
plt.title('Correlation Matrix', color='black', fontsize=20)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.show()

corr_unstacked = corr.unstack()
corr_unstacked = corr_unstacked[corr_unstacked.index.get_level_values(0) != corr_unstacked.index.get_level_values(1)]

# Get the top 10 correlations by absolute value
top_10_corr = corr_unstacked.sort_values(ascending=False).head(10)

# Print the top 10 correlations with their actual values
top_10_corr_with_values = [(index, corr_unstacked[index]) for index in top_10_corr.index]
print("Top 10 correlations:")
for (col_pair, value) in top_10_corr_with_values:
    print(f"{col_pair}: {value}")
```

### Bar Plots for Visualizing Distribution of Data Across Different Classes (Categorical Features)

```python
wilderness_columns = [col for col in df.columns if 'Wilderness_Area' in col]
df_wilderness = df[wilderness_columns]
df_wilderness_counts = df_wilderness.apply(pd.Series.value_counts).fillna(0)

ax = df_wilderness_counts.plot(kind='bar', stacked=True, figsize=(16,8))
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
ax.ticklabel_format(style='plain', axis='y')

plt.title('Distribution of Wilderness Area', color='black', fontsize=16)
plt.xlabel('Wilderness Area', color='black', fontsize=14)
plt.ylabel('Count', color='black', fontsize=14)
plt.tick_params(axis='both', which='major', labelsize=14)
plt.xticks(rotation=360)  
plt.grid(True, linewidth=1, color='black', alpha=0.2)
plt.legend(fontsize=14, ncol=1, loc="upper right")
plt.show()
```

```python
soiltype_columns = [col for col in df.columns if 'Soil_Type' in col]

# Splitting the soiltype_columns list into chunks of 10
chunks = [soiltype_columns[i:i + 10] for i in range(0, len(soiltype_columns), 10)]

# Plotting each chunk
for i, chunk in enumerate(chunks):
    df_soiltype = df[chunk]
    df_soiltype_counts = df_soiltype.apply(pd.Series.value_counts).fillna(0)

    ax = df_soiltype_counts.plot(kind='bar', stacked=True, figsize=(16, 8))
    ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))
    ax.ticklabel_format(style='plain', axis='y')

    plt.title(f'Distribution of Soil Type (Part {i+1})', color='black', fontsize=16)
    plt.xlabel('Soil Type', color='black', fontsize=14)
    plt.ylabel('Count', color='black', fontsize=14)
    plt.tick_params(axis='both', which='major', labelsize=14)
    plt.xticks(rotation=360)
    plt.grid(True, linewidth=1, color='black', alpha=0.2)
    plt.legend(fontsize=14, ncol=2, loc="upper right")
    plt.show()
```
